# Import dependencies

```python
import gym 
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
import os
```

# Load enviroment

```python
# The enviroment we using
env = gym.make('LunarLander-v2')
```

```python
episodes = 5
for episode in range(episodes):
    state = env.reset()
    done = False
    total_reward = 0
    
    while not done:
        # Optionally, render the environment to visualize the episode
        env.render()
        # Take a random action from the action space
        action = env.action_space.sample()
        # Take a step in the environment based on the selected action
        state, reward, done, info = env.step(action)

        total_reward += reward
        
    # After the episode is finished, print the total reward obtained
    print(f"Episode {episode + 1} - Total Reward: {total_reward}")
env.close()
```

# Train the model

```python
def train_ppo():
    # Create the Lunar Lander environment
    env = gym.make('LunarLander-v2')

    # Create the PPO agent
    model = PPO('MlpPolicy', env, verbose=1)

    # Train the agent
    model.learn(total_timesteps=500000)

    # Save the trained model
    model.save('ppo_lunar_lander500k')

    # Close the environment
    env.close()
```

```python

if __name__ == "__main__":
    train_ppo()
```

# Test model

```python

def test_model():
    # Create the Lunar Lander environment
    env = gym.make('LunarLander-v2')

    # Load the trained PPO agent
    model = PPO.load('ppo_lunar_lander500k')

    # Evaluate the agent
    eval_episodes = 5
    for episode in range(eval_episodes):
        obs = env.reset()
        done = False
        total_reward = 0.0

        while not done:
            # Render the environment
            env.render()

            # Get the action from the model
            action, _ = model.predict(obs, deterministic=True)

            # Take the action in the environment
            obs, reward, done, _ = env.step(action)

            # Accumulate the total reward
            total_reward += reward

        # Print the reward for each episode
        print(f"Episode {episode + 1}: Reward = {total_reward}")

    # Close the environment
    env.close()


```

```python
# Call the test_model function
test_model()
```
